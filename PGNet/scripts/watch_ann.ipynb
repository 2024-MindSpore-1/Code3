```python
import cv2
import numpy as np
import json
import matplotlib.pyplot as plt
import os.path as osp
```

# totaltext

```python
annfilepath = '../data/total_text/train/train.txt'
img_root = '../data/total_text/train/'
total_anns = {}
with open(annfilepath) as f:
    for line in f.readlines():
        filename, data = line.split('\t')
        total_anns[filename] = json.loads(data)
```

```python
img_name = 'rgb/img19.jpg'
image = cv2.imread(osp.join(img_root, img_name))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
anns = total_anns[img_name]
```

```python
draw_img = image.copy()
H, W, C = draw_img.shape
r = max(1, min(H, W) // 100)

for ann in anns:
    points = np.array(ann['points']).astype(int)
    cv2.polylines(draw_img, [points], True, (0, 0, 255), thickness=2)
    for pt in points:
        cv2.circle(draw_img, pt, r, (255, 0, 0), -1)
plt.imshow(draw_img)
```
